```python
# Grouping & Aggregation

import pandas as pd

data = {
    "Name": ["Swaraj", "Nikita", "Abhijit", "Jayed", "Aadesh"],
    "Subject": ["Math", "Science", "Math", "Science", "Math"],
    "Marks": [85, 78, 92, 88, 76],
    "Gender": ["M", "M", "F", "F", "M"]
}

df = pd.DataFrame(data)

print(df)

```

```python
# Group by Subject and calculate average marks
avg_marks = df.groupby("Subject")["Marks"].mean()

print("Average Marks by Subject:")
print(avg_marks)
```

```python
# Group by Gender and count the number of students
count_students = df.groupby("Gender")["Name"].count()

print("Number of Students by Gender:")
print(count_students)
```

```python
# Group by Subject and calculate total marks
total_marks = df.groupby("Subject")["Marks"].sum()

print("Total Marks by Subject:")
print(total_marks)
```

```python
# Group by Subject and calculate multiple aggregations
agg_result = df.groupby("Subject")["Marks"].agg(["mean", "max", "min", "count"])

print("Aggregated Results:")
print(agg_result)
```
