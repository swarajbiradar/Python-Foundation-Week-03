```python
# Grouping and Aggregation Example(Titanic Dataset)

import pandas as pd

titanic = pd.DataFrame({
    "Class": ["1st", "2nd", "3rd", "1st", "2nd", "3rd"],
    "Gender": ["M", "F", "M", "F", "M", "F"],
    "Survived": [1, 1, 0, 1, 0, 1]
})

print(titanic)
```

```python
# Group by Class and calculate the total number of survivors
survival_count = titanic.groupby("Class")["Survived"].sum()

print("Survival Count by Class:")
print(survival_count)
```

```python
# Group by Class and calculate the survival rate
survival_rate = titanic.groupby("Class")["Survived"].mean()

print("Survival Rate by Class:")
print(survival_rate)
```
