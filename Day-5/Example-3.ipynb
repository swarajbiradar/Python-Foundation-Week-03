```python
# Handling Duplicate Columns

import pandas as pd

df_a = pd.DataFrame({
    "ID": [1, 2],
    "Value": [100, 200]
})

df_b = pd.DataFrame({
    "ID": [1, 2],
    "Value": [300, 400]
})

merged = pd.merge(df_a, df_b, on="ID", suffixes=("_A", "_B"))

print("Handling Duplicate Columns:")
print(merged)
```
