```python
import pandas as pd

# Create a Series from a list of fruits
fruits = ["avocado", "papaya", "grapes", "pineapple"]

series_fruits = pd.Series(fruits)

print("Fruits Series:")
print(series_fruits)

# Create a Series from a list of exam scores with custom indices
scores = [85, 90, 78, 92]

series_scores = pd.Series(scores, index=["A", "B", "C", "D"])

print("\nExam Scores Series:")
print(series_scores)
```
