```python
import pandas as pd
# Example: Student Information

data = {
    "Name": ["Swaraj", "Jayed", "Pramod"],
    "Age": [20, 21, 19],
    "Grade": ["A", "B", "A"]
}

df = pd.DataFrame(data)

print("\nStudent DataFrame:")
print(df)
print("-------------------------------")
# View first rows
print("\nHead:")
print(df.head())
print("-------------------------------")
# View last rows
print("\nTail:")
print(df.tail())
print("-------------------------------")
# Summary statistics
print("\nDescribe:")
print(df.describe())
print("-------------------------------")
# Check data types
print("\nData Types:")
print(df.dtypes)
```
