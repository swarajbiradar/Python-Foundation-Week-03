```python
import pandas as pd

data = {
    "Name": ["Swaraj", "Jayed", "Aadesh", "Sruthi", "kaif"],
    "Marks": [85, 72, 90, 65, 88],
    "Age": [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)

print(df)

# Get row with index 2
print(df.loc[2])

# Get specific columns
print(df.loc[:, ["Name", "Marks"]])

# Get first row
print(df.iloc[0])

# Get rows 1 to 3 and columns 0 to 1
print(df.iloc[1:4, 0:2])

# Get rows where Marks > 80
high_scores = df[df["Marks"] > 80]
print("Students with marks > 80:")
print(high_scores)

filtered = df[(df["Marks"] > 80) & (df["Age"] < 21)]

print("Marks > 80 and Age < 21:")
print(filtered)

subset = df[1:4]

print("Subset of rows:")
print(subset)
```
