```python
import pandas as pd

titanic = pd.DataFrame({
    "Name": ["Swaraj", "Hariom", "Nikhil", "Abhijit"],
    "Age": [22, 38, 26, 35],
    "Class": ["3rd", "1st", "3rd", "1st"],
    "Survived": [0, 1, 1, 1]
})

print(titanic)

# Filtering rows where Age is greater than 30
print("Passengers age > 30:")
print(titanic[titanic["Age"] > 30])

# Filtering rows where Class is '1st'
print("1st Class Passengers:")
print(titanic[titanic["Class"] == "1st"])

# Filtering rows where Survived is 1
print("Survived Passengers:")
print(titanic[titanic["Survived"] == 1])
```
