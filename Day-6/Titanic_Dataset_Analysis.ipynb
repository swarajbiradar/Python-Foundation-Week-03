```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print(df.info())
print(df.describe())

print(df.head())
```

```python
# Basic Data Exploration
print("Columns:")
print(df.columns)

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

```python
# Checking for Missing Values
print("Missing Values:")
print(df.isnull().sum())
```

```python
older_passengers = df[df["Age"] > 30]

print(older_passengers.head())

female_passengers = df[df["Sex"] == "female"]

print(female_passengers.head())
```

```python
# Grouping & Aggregation
# Survival rate by gender
survival_gender = df.groupby("Sex")["Survived"].mean()

print("Survival Rate by Gender:")
print(survival_gender)
```

```python
# Survival rate by class
survival_class = df.groupby("Pclass")["Survived"].mean()

print("Survival Rate by Class:")
print(survival_class)
```

```python
# Average age of survivors
avg_age_survivors = df[df["Survived"] == 1]["Age"].mean()

print("Average Age of Survivors:", avg_age_survivors)
```

```python
# Filtering Data
filtered = df[(df["Sex"] == "female") & (df["Survived"] == 1)]

print("Female Survivors:")
print(filtered.head())
```
